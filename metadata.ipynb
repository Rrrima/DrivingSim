```python
# TORlit:4s,7s
# traffic density: 0-low, 1-high
# workload: 0-low, 1-high
meta_dict = {
    'A': [4,0,0],
    'B': [7,0,0],
    'C': [4,1,0],
    'D': [7,1,0],
    'E': [4,0,1],
    'F': [7,0,1],
    'G': [4,1,1],
    'H': [7,1,1],
}
event_list = ['bicycle_lc','cons_onleft_lk','cons_ahead_lc','sens_error_lk','car_ahead_lc','no_mark_cv','sens_error_cv','car_shoulder_lc']
```

```python
def query_by_subject(sub_id,eid):
    situation_id = meta_data[meta_data['subject']==sub_id].values[0][1:][eid]
    return meta_dict[situation_id[0]]
```
