```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns', 500)
pd.options.display.max_rows = 999
```

```python
result1 = pd.read_csv('result_workload_carahead.csv')
res1 = result1.drop([7,44,52,4])
result2 = pd.read_csv('result_workload_bicycle.csv')
res2 = result2.drop([78])
```

```python
result = res1.append(res2)
```

```python
plt.plot(result.brake,result.ttc_time,'o')
```

```python
from scipy.stats import pearsonr
stat, p = pearsonr(result.brake,result.ttc_time)
print('Statistics=%.3f, p=%8f' % (stat, p))
```
