```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns', 500)
pd.options.display.max_rows = 999
```

```python
result = pd.read_csv('result_workload_bicycle.csv')
```

```python
# drop outlier
res = result.drop([78])
```

```python
plt.plot(res.brake,res.ttc_time,'o')
plt.grid()
```

```python
plt.plot(res.steer,res.ttc_time,'o')
```

```python
plt.plot(res.tot,res.ttc_time,'o')
```

```python
plt.plot(res.gas,res.ttc_time,'o')
```

```python
res.boxplot(column=['brake', 'steer','gas','tot'])
```

```python
from scipy.stats import ttest_ind
stat, p = ttest_ind(res['gas'], res['steer'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
```
