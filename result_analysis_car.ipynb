```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns', 500)
pd.options.display.max_rows = 999
```

```python
result = pd.read_csv('result_workload_carahead_addlane.csv')
```

```python
# # drop outlier
res = result.drop([7,44,52,4])
```

```python
plt.figure(figsize=(6,6))
plt.plot(res.brake,res.ttc_time,'ok',alpha=0.6)
plt.xlim(0,8)
plt.ylim(0,8)
plt.plot([0,8],[0,8],'k--',alpha=0.5)
plt.xlabel('Brake Reaction Time')
plt.ylabel('TTC calculate point')
```

```python
plt.figure(figsize=(6,6))
plt.plot(res.steer,res.ttc_time,'ok',alpha=0.6)
plt.xlim(0,8)
plt.ylim(0,8)
plt.plot([0,8],[0,8],'k--',alpha=0.5)
plt.xlabel('Steer Reaction Time')
plt.ylabel('TTC calculate point')
```

```python
plt.figure(figsize=(6,6))
plt.plot(res.gas,res.ttc_time,'ok',alpha=0.6)
plt.xlim(0,8)
plt.ylim(0,8)
plt.plot([0,8],[0,8],'k--',alpha=0.5)
plt.xlabel('Throttle Reaction Time')
plt.ylabel('TTC calculate point')
```

```python
plt.figure(figsize=(6,6))
plt.plot(res.tot,res.ttc_time,'o',alpha=0.6)
plt.xlim(0,8)
plt.ylim(0,8)
plt.plot([0,8],[0,8],'k--',alpha=0.5)
plt.xlabel('TOT')
plt.ylabel('TTC calculate point')
```

```python
res.boxplot(column=['brake', 'steer','gas','tot'])
```

```python
from scipy.stats import ttest_ind
stat, p = ttest_ind(res['brake'], res['gas'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
```
